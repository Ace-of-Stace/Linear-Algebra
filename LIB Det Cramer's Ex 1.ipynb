#### Introduction 

Gabriel Cramer was a mathematician born July 31, 1704 in Geneva, Switzerland. [1,2] He revolutionized mathematics instruction and contributed greatly to math theory in his time.  Collaboration with leading mathematicians such as Johann Bernoulli, Leonhard Euler, Edmond Halley, Abraham de Moivre and James Stirling allowed him to develop influential methodology, including a technique of solving systems of equations for an arbitrary number of unknowns, published in 1750. [3,4,5,6,7] This procedure eventually became known as Cramer’s Rule. 
Cramer's Rule is a method for solving systems of linear equations which can be used when the system is nonsingular or has a nonzero determinant.[8] The solution is found by finding the determinant of the weighted matrix. The determinant can be thought of as a function whose input is a square matrix and whose output is a single number that conveys information about the area a vector space resides in.[9] While typically Gaussian elimination is used as a means for solving such systems, Cramer’s Rule has a specific advantage: able to solve for a single coordinate of $x$ in $Ax=b$ without having to solve the entire system.  However, holistically Cramer’s Rule is computationally inefficient and is limited since it cannot be used when the determinant is equal to zero.

#### An Example

#### To apply Cramer's Rule to a system of two equations, the set up is as follows:

$
\Large
\begin{cases}
\begin{equation} ax+by=\boxed{c}\\
dx+ey=\boxed{f}\\
\end{equation}
\end{cases}
$

#### The coefficient matrix's determinant, called D is:
$
\Large
D =
\begin{bmatrix}
a & b\\
d & e\\
\end{bmatrix}
$
#### And the solution column is:
$
\Large
\begin{bmatrix}
\boxed{c}\\
\boxed{f}\\
\end{bmatrix}
$

#### Cramer's rule is applied by substituting the solution column values into the coefficient matrix column values for the variable of interest and then evaluating the determinant:

$
\Large
\begin{equation} x =
\frac{
\begin{bmatrix}
\boxed{c} & b\\
\boxed{f} & e\\
\end{bmatrix}}
{\begin{bmatrix}
a & b\\
d & e\\
\end{bmatrix}}\\
\\
y =
\frac{
\begin{bmatrix}
a & \boxed{c}\\
d & \boxed{f}\\
\end{bmatrix}}
{\begin{bmatrix}
a & b\\
d & e\\
\end{bmatrix}}
\end{equation}
$

#### So, let's apply this to an example:
$
\Large
\begin{cases}
\begin{equation} 2x+3y=\boxed{7}\\
3x+y=\boxed{5}\\
\end{equation}
\end{cases}
$


$
\Large
\begin{equation} x =
\frac{
\begin{bmatrix}
\boxed{7} & 3\\
\boxed{5} & 1\\
\end{bmatrix}}
{\begin{bmatrix}
2 & 3\\
3 & 1\\
\end{bmatrix}}
=
\frac{7-15}{2-9}=
\frac{-8}{-7}=\boxed{1.142}
\end{equation}\\
$


$
\Large
\begin{equation} y =
\frac{
\begin{bmatrix}
2 & \boxed{7}\\
3 & \boxed{5}\\
\end{bmatrix}}
{\begin{bmatrix}
2 & 3\\
3 & 1\\
\end{bmatrix}}
=
\frac{10-21}{2-9}=
\frac{-11}{-7}=\boxed{1.571}
\end{equation}
$

#### In this example, $x=1.142$ and $y=1.571$ are the solutions to the system of equations, but if we were only interested in the x-value, we could have done the first half of the problem only. It is this scenario when Cramer's Rule becomes very useful.

#### The Final Code

The cohort code (see Appendix A) built a coefficient solving tool integrated with a component to measure the time required by the program to solve a system of equations via “Cramer’s Rule” matrix. The code allowed the size of the matrix (S) to be controlled by the user and used a random integer generator to create a square $n x n$ array with those specifications.  The code contained stipulations that guaranteed the determinant$(det)$ was linearly independent, $(det≠ 0)$, since Cramer’s Rule is inapplicable in that case. The code then generated a matrix, $A$, found the inverse of that matrix, $Ai$, and solved for the determinant. Again, arguments were coded in to keep only nonzero solutions. Then a randomly generated B matrix was created to replace the columns of the A matrix and the $det(A)$, $det (Ai)$ and variable, $xi$ were calculated while being monitored by a time keeping algorithm.

#### Conclusion

Linear algebraic system of equations can be solved several ways.  Cramer’s rule is one method that works well for small systems, or when the entire system doesn’t need to be solved. However, as shown above, as unknowns are added to the matrix, processing time increases exponentially, and the method becomes computationally expensive. These “expensive” programs are those which require more resources, such as time, space, arithmetic operations, etc. For example, time complexity for algorithms can be expressed as a growth rate function of the size of the input (represented by $Ο(n)$).[10] The time complexity function for Cramer’s Rule is commonly described as $O(n!)$, which would make it impractical, computationally.[11]
